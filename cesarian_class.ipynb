```r
mkcat=function(x,val) {
    y=c()
    y.length=length(x)
    count=1
    for(i in x){
        #print(i)
        #print(count)
        if( i == val) { y[count] = 1}
        else{ y[count] = 0}
        count=count+1
    }
    return (y)
}
```
